```python

def hist(image):
    dico ={}
  
    for i in range(len(image)):
        for j in range(i):
            if image[i][j] in dico:
                dico[image[i][j]]+=1
            else:
                dico[image[i][j]]=1

  
    plt.bar(dico.keys(),dico.values())
    plt.show()
```

```python
hist(data_rice)
```

```python
hist(data_spine)
```

```python
def cumulative_hist(image):
    dico ={}
    for i in range(len(image)):
            for j in range(i):
                if image[i][j] in dico:
                    dico[image[i][j]]+=1
                else:
                    dico[image[i][j]]=1

    cumulative_dico = {}
    current_sum = 0

    sorted_keys = sorted(dico.keys())

    for key in sorted_keys:
        current_sum += dico[key]
        cumulative_dico[key] = current_sum

    plt.bar(cumulative_dico.keys(), cumulative_dico.values())

    plt.show()
```

```python
cumulative_hist(data_spine)
```

```python
cumulative_hist(data_rice)
```

```python
import numpy as np

def recadrer_dynamique(image, a=None, b=None):
    """
    Recadre la dynamique de l'image entre les valeurs a et b.
    Si a et b ne sont pas spécifiés, on utilise les valeurs min et max de l'image.

    Parameters:
    - image : np.ndarray, image en niveaux de gris ou en couleur
    - a : float, valeur minimale de la dynamique souhaitée
    - b : float, valeur maximale de la dynamique souhaitée

    Returns:
    - image_recadree : np.ndarray, image avec la dynamique recadrée
    """
    if a is None:
        a = np.min(image)
    if b is None:
        b = np.max(image)

    image_recadree = (image - a) * (255 / (b - a))
    
    image_recadree = np.clip(image_recadree, 0, 255)

    return image_recadree.astype(np.uint8)
```

```python
image_recadree1 = recadrer_dynamique(data_spine)
plt.imshow(image_recadree1, cmap='gray')
plt.axis('off') 
plt.title("Image avec dynamique recadrée")
plt.show()
```
